## Recursion and functions
> What stops us from using iteration in the main body of the program instead of recursive functions? Do we strive for the fewest lines of code in the main body, while having all the code in function form?

No. It's totally fine to have iteration instead of recursion. But some problems are much easier to write recursively than iteratively. We haven't seen many of those problems yet. In particular **tree operations** and some **graph operations** are easy to write with recursion and harder to write iteratively.

Some languages *only* have recursion and don't have iteration at all. This is generally a pain.

> How are variables defined in loops local if they can be accessed in the global scope?

They have the scope of their definition. So they will are local to the current scope (if that's global, it will be global). They don't define a *new* scope, but inherit the scope where they are defined.

> Can recursion only involve one function, or can it involve several in a cycle?

There's no restriction. This is **mutual recursion** and is occasionally used. It can be hard to track what is going on unless you are very disciplined.  There is an example in the C part of the lab.

### Did not quite understand the global variables

```python
# not inside a function here
# this variable will be global
# ALL functions can see it
a_global = 5

def a_function():
    # a_local is only visible inside this function
    a_local =  a_global * 10
    
    print(a_local)
    # after here, a_local evaporates into the ether
    
def another_function():
    another_local = a_global * 20    
    print(another_local)
    
a_function()
another_function()
```

```python
def bad_function():
    a_global = 20
    # WARNING -- this just creates a new local variable a_global and temporarily
    # hides the global one
    
def good_function():
    global a_global
    a_global= 2000
    # This now sets the value of a_global
    
a_global = 100
bad_function() # does nothing 
print(a_global)
good_function()
print(a_global)
```

## Code criticism: Don't Repeat Yourself (DRY)
One very easy to recognise indicator of poor code quality is **repetition**. Code shouldn't be repeated. Copy-and-paste code is a sign of a hack job. Repetition should be coded as iteration (if possible) or if the problem doesn't require *exact* repetition, the code should be packaged up into functions "factor out" the repeated code.

#### Why no Ctrl-C / Ctrl-V?
Why is this bad? Repetition means that you have different parts of the program doing very similar things. What if you wanted to change the behaviour of all the code consistently (e.g. to fix a bug or add a feature)? You'd have to go through each repetition and change or fix it. 

What if one copy of a bit of code was very slightly different? It's very hard to recognise slight differences when when things are repeated, because the human perceptual system filters the variation out.

**Repeated code is inflexible, fragile, hard to read and hard to keep consistent.** It makes your life harder, and it indicates that the problem hasn't been thought out about enough to boil it down into simple, reusable components.

Let's look at some bad code:

```python

pool_len = 50
pool_depth = 2
pool_width = 16
chlorine_percent = 0.01

def compute_chlorine(pool_len, pool_depth, 
                     pool_width, chlorine_percent):
    pool_volume = pool_len * pool_depth * pool_width
    pool_chlorine = (chlorine_percent/100.0) * pool_volume
    return pool_chlorine
    
pool_chlorine = compute_chlorine(pool_len, pool_depth, 
                                 pool_width, chlorine_percent)
print("Chlorine needed %.2fl" % (pool_chlorine))
```

## Q3 Scope
All of the variables in this function `mod_code()` are global. Which of them need to be for the function to make sense? Assume that `mod_den` cannot return values.

```python
accumulated_string = ""
num = 32
den = num-1
char = "*"

def mod_code():
    global num, den, char, accumulated_string
    if num%den==0:
        accumulated_string += char     
    den = den - 1
    accumulated_string += "-"

while den>1:
    mod_code()
print(accumulated_string)
```
